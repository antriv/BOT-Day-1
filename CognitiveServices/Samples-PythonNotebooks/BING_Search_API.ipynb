```python
data = response.read().decode("utf-8")
CONFIG = json.loads(data)
subscription_key = CONFIG['subscription_key_Bing_Search']
```

```python
########### Python 3.2 #############
# grab key from json in config
headers = {
    # Request headers
    'Ocp-Apim-Subscription-Key': subscription_key
}

# Search for top three microsoft webpages
params = urllib.parse.urlencode({
    # Request parameters
    'q': 'microsoft',
    'count': '3',
    'offset': '0',
    'mkt': 'en-us',
    'safeSearch': 'Moderate',
})

try:
    conn = http.client.HTTPSConnection('api.cognitive.microsoft.com')
    conn.request("GET", "/bing/v5.0/news/search?%s" % params, "{body}", headers)
    response = conn.getresponse()
    data = response.read()
    print(data)
    
    conn.close()
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))
```

```python
## Search news with a particular string
import requests

def bing_search(query):
    url = 'https://api.cognitive.microsoft.com/bing/v5.0/search'
    # query string parameters
    payload = {'q': query}
    # custom headers
    headers = {'Ocp-Apim-Subscription-Key': subscription_key}
    # make GET request
    r = requests.get(url, params=payload, headers=headers)
    # get JSON response
    return r.json()

j = bing_search('Barak Obama')
print(j.get('webPages', {}).get('value', {}))
```
